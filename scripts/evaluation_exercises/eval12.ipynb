12.2. Modelo de NN para o dataset breast-bin.csv

```python
# imports
import numpy as np
from si.io_package.data_file import read_data_file
from sklearn.preprocessing import StandardScaler
from si.io_package.csv_file import read_csv
from si.metrics.cross_entropy import cross_entropy
from si.model_selection.split import train_test_split
from si.neural_networks.layers import Dense, ReLUActivation, LinearActivation, SigmoidActivation, SoftMaxActivation
from si.neural_networks.nn import NN
```

```python
# dataset
filename_bb = "C:/Users/Ana/Documents/GitHub/mbioinf-sib/datasets/breast-bin.data"
bb_dataset = read_data_file(filename_bb, sep=",", label=True)
```

```python
# standardization
bb_dataset.X = StandardScaler().fit_transform(bb_dataset.X)
```

```python
# split dataset
bb_train, bb_test = train_test_split(bb_dataset)
```

```python
bb_train.shape()
```

```python
bb_l1 = Dense(input_size=9, output_size=9)
bb_l2 = Dense(input_size=9, output_size=5)
bb_l3 = Dense(input_size=5, output_size=1)

bb_l1_relu = ReLUActivation()
bb_l2_relu = ReLUActivation()
bb_l3_sg = SigmoidActivation()

# layers
layers = [bb_l1, bb_l1_relu, bb_l2, bb_l2_relu, bb_l3, bb_l3_sg]

# NN
bb_model = NN(layers=layers)
```

```python
bb_model.predict(dataset=bb_train)
```

# 3. Testing clustering module

```python
from si.io_package.csv_file import read_csv
from si.clustering.kmeans import KMeans
from si.decomposition.pca import PCA
import numpy as np
import matplotlib.pylab as plt
```

```python
# datasets
filename = "C:/Users/Ana/Documents/GitHub/mbioinf-sib/datasets/iris.csv"
iris = read_csv(filename, sep=",", label=True, features=True)
```

```python
# kmeans
k = 3
kmeans = KMeans(k)
distances = kmeans.fit_transform(iris)
predictions = kmeans.predict(iris)
predictions
```

```python
# plotting
for i in range(k):
    k_sample = iris.X[np.where(predictions == i)]
    plt.plot(k_sample[:, 0], k_sample[:, 1], 'o')
    lines = plt.plot(kmeans.centroids[i, 0], kmeans.centroids[i, 1], 'kx')
    plt.setp(lines, ms=15.0)
    plt.setp(lines, mew=2.0)

plt.show()
```

```python
# pca
pca = PCA(n_components=2)
dataset_pca = pca.fit_transform(dataset=iris)
dataset_pca
```

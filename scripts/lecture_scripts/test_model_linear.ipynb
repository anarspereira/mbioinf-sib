Ridge Regression

```python
# imports
from si.io_package.csv_file import read_csv
from si.model_selection.split import train_test_split
from si.linear_model.ridge_regression import RidgeRegression
from si.linear_model.logistic_regression import LogisticRegression
```

```python
# datasets
filename_cpu = "C:/Users/Ana/Documents/GitHub/mbioinf-sib/datasets/cpu.csv"
cpu_dataset = read_csv(filename_cpu, sep=",", label=True, features=True)
```

```python
# standardize data
from sklearn.preprocessing import StandardScaler
cpu_dataset.X = StandardScaler().fit_transform(cpu_dataset.X)
```

```python
# split dataset
train_dataset, test_dataset = train_test_split(cpu_dataset)
test_dataset.y
```

```python
# ridge regressor
ridge = RidgeRegression(max_iter=2000)
ridge.fit(train_dataset)
predictions = ridge.predict(test_dataset)
predictions
```

```python
ridge.score(test_dataset)
```

```python
ridge.cost(test_dataset)
```

```python
ridge.cost_history
```

```python
ridge.plot_cost_history()
```
